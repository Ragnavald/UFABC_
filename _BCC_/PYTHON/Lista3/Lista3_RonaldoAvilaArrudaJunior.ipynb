# Exercício 1: polinômio
Peça para o usuário digitar um número real $a$ e um número real $b$. 

Você pode assumir que o usuário irá digitar números tais que $a<b$.

Faça o gráfico da função $f(x) = x^4 - 10x^3-100x^2-1000x$ para x entre $a$ e $b$ (não incluso).

Use incrementos de 0.2.

Para ajudar você a conferir sua resposta, aqui está o gráfico obtido para $a = -10$ e $b = 10$.

image.png

```
print("Digite um valor para a e b, assumindo que a<b")
a = float(input("Digite a:"))
b = float(input("Digitr b:"))
x = np.arange(a,b,0.2)
y = x**4-10*x**3-100*x**2-1000*x
plt.plot(x,y,'')
```

# Exercício 3: funções trigonométricas

Faça o gráfico da função $f(x) = \text{sen}(x^2) + \text{cos}(x)\cdot\text{cos}(x)$ para x entre $-6$ e $6$ (não incluso) com incrementos de 0.1

O seu gráfico deve ficar assim:

image.png


```
x = np.arange(-6,6,0.1)
y = np.sin(x**2)+np.cos(x)*np.cos(x)
plt.plot(x,y,'')
```

# Exercício 4: espiral


Faça um programa que pede para o usuário digitar um valor real $c$ entre 1 e 2. 

Você pode assumir que o usuário digitará corretamente. Após ler o valor de c:

- Crie uma lista t de números começando em 0 e terminando em 50 (não incluso) com incrementos de 0.5.

- Crie $x = t\cdot \text{cos}(t^c)$

- Crie $y = t\cdot \text{sen}(t^c)$

- Faça o gráfico de $x$ e $y$ (com ponto-linha ligando os pontos).

Use o comando `plt.axis("equal")` para que a escala dos eixos fique igual.

Para ajudar você a conferir sua resposta, aqui está o gráfico com $c=1$:

image.png



```
c = float(input("Digite um valor entre 1 e 2:"))
t = np.arange(0,50,0.5)
x = t*np.cos(t**c)
y = t * np.sin(t**c)
plt.axis("equal")
plt.plot(x,y,".-")
```
